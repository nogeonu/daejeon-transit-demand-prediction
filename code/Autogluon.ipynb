```python
from sklearn.metrics import mean_absolute_error
from autogluon.tabular import TabularDataset, TabularPredictor
import pandas as pd
```

### DATA 불러오기

```python
train = pd.read_csv('C:/python/merged_data (1).csv')
test = pd.read_csv('C:/python/merged_test (1).csv')

display(train.head(3), test.head(3))
```

### model 생성 및 train_data 학습

```python
# Autogluon을 사용하여 모델 학습
predictor = TabularPredictor(
    label='RIDE_DEMAND', problem_type='regression', eval_metric='mae'
).fit(train)
```

## Train_set MAE 값 확인하기

```python
# 훈련 데이터에 대한 예측값 생성 및 MAE 계산 
train_predictions = predictor.predict(train.drop('RIDE_DEMAND',axis=1))
train_target = train['RIDE_DEMAND']  # 훈련 데이터의 실제 타겟 값

train_mae = mean_absolute_error(train_target, train_predictions)
print(f"Training set MAE: {train_mae}")
```

### Autogluon_model 리더보드 생성

```python
leaderboard = predictor.leaderboard(train, silent=True)
leaderboard
```

### 제출용 submisson.csv 생성

```python
# 테스트 데이터에 대한 예측값 생성
test_predictions = predictor.predict(test)
```

```python
submission = test
```

```python
# 테스트 데이터에 예측값 열 추가
submission['RIDE_DEMAND'] = test_predictions

# CSV 파일로 저장
submission.to_csv('submission.csv', index=False)
```
